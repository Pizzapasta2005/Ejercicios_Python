```python
filas_nulas = datos[datos.isnull().any(axis=1)]

filas_nulas
```

```python
def filtrar_tipo_fuego(data):
  return data[(data["Type 1"] == "Fire") | (data["Type 2"] == "Fire")]

data_sin_nulos = datos.dropna()
data_fuego = filtrar_tipo_fuego(data_sin_nulos)
display(data_fuego.head(20))
```

```python

promedio_defensa = data_fuego["Def"].mean()
print(f"Promedio de defensa de los pokemon de tipo fuego: {promedio_defensa:.2f}")
```

```python
def filter_fire(datos):
  return datos[(datos["Type 1"] == "Fire") | (datos["Type 2"] == "Fire")]

data_fire = filter_fire(datos)
print("\nPrimeros 20 Pokemon de tipo fuego:")
display(data_fire.head(20))
```

```python
top10 = data_fire.sort_values("Att", ascending=False).head(10)
plt.figure(figsize=(19,5))
plt.bar(top10["Name"], top10["Att"])
plt.title("Top 10 Pokemones de tipo fuego segun su nivel de ataque")
plt.xlabel("Nombre Pokemons")
plt.ylabel("ataque")
plt.tight_layout()
plt.show
```
